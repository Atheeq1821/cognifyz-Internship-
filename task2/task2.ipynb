```python
#imprting the libraries and dataset
import pandas as pd
task2=pd.read_csv("cleaned1.csv")
rows,columns=task2.shape
task2.head()
```

```python
#Column Names
task2.columns
```

```python

featureList=task2[[ 'Has Table booking','Cuisines',
       'Has Online delivery', 'Is delivering now','Price range', 'Aggregate rating', 'Rating text',  'Rating color',
       'Votes']].fillna("")
```

```python
#Content Based Filtering
y=task2['Restaurant Name']
x_text=featureList['Cuisines']+" "+featureList[ 'Rating color']+" "+featureList['Rating text']+" "+ str(featureList['Votes']) +" "+ str(featureList['Price range'])

from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer()
x=tfid.fit_transform(x_text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarityScore= cosine_similarity(x)
similarityScore
```

```python
#User Input 
userInput = input("Enter Your Preferred Cuisines :")
cuisineList= task2["Cuisines"]
for i in range(len(cuisineList)):
    if userInput in cuisineList[i]:
        k=task2.iloc[i]["Restaurant Name"]
        index=i
        break
RecommendationScore=list(enumerate(similarityScore[index]))  


        
```

```python
#Recommendation of Restaurants
sortedRestaurantList=sorted(RecommendationScore,key=lambda x:x[1],reverse=True)
count=0
i=0
for rest in sortedRestaurantList:
    Index = rest[0]
    k=task2.iloc[Index]["Restaurant Name"]
    print(k)
    count+=1
    if count==10:
      break
    i+=1
```
